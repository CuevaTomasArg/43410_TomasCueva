```python
# Nombre de la tabla en la base de datos
table = 'dishes_per_sale'

# Genera el script SQL
script_sql = ''
for _, row in df_dishes_per_sale.iterrows():
    # Obtiene los valores de cada columna
    valores = ', '.join([f"'{str(value)}'" for value in row])
    
    # Genera el comando INSERT para cada fila
    script_sql += f"INSERT INTO {table} VALUES ({valores});\n"

# Guarda el script SQL en un archivo
with open('./script_entregable/script_insert_dishes_per_sale.sql', 'w', encoding = 'utf-8') as file:
    file.write(script_sql)

print('Script SQL generado con Ã©xito.')
```
